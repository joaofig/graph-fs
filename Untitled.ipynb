```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
g = nx.DiGraph()
```

```python
edge_list = [
    ('A', 'B', 0.9),
    ('A', 'C', 0.3),
    ('A', 'D', 0.5), 
    ('A', 'Y', 0.5),

    ('B', 'A', 0.7),
    ('B', 'C', 0.3),
    ('B', 'D', 0.3),
    ('B', 'Y', 0.9),

    ('C', 'A', 0.2),
    ('C', 'B', 0.1),
    ('C', 'D', 0.1),
    ('C', 'Y', 0.7),

    ('D', 'C', 0.2),
    ('D', 'B', 0.1),
    ('D', 'A', 0.3),
    ('D', 'Y', 0.6),
    
    ('Y', 'A', 1e-5),
    ('Y', 'B', 1e-5),
    ('Y', 'C', 1e-5),
    ('Y', 'D', 1e-5),    
]
```

```python
g.add_weighted_edges_from(edge_list)
```

```python
nx.draw_networkx(g)
```

```python
page_rank = nx.pagerank_numpy(g)
```

```python
total_fw = sum([val for key, val in page_rank.items() if key != 'Y'])
```
